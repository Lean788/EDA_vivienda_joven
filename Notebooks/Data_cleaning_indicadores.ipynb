```python
# importaciones básicas
import numpy as np
import pandas as pd
```

```python
indicadores = pd.read_csv('../data/dataset8_indicadores.csv', sep=';', encoding='utf-8')
indicadores.head(25)
```

```python
indicadores.drop([0,1], inplace=True)
```

```python
indicadores.columns = indicadores.iloc[0]
```

```python
indicadores.reset_index(drop=True ,inplace=True)
```

```python
indicadores.drop([0,1,2], inplace=True)
```

```python
indicadores.dropna(thresh=9,inplace=True)
```

```python
indicadores.iloc[1] = indicadores.iloc[1].str.replace(',','.')
```

```python
indicadores.reset_index(drop=True ,inplace=True)
```

```python
indicadores['2011'] = indicadores['2011'].str.replace(',','.').astype(float)
```

```python
indicadores.to_csv('../data/clean/indicadores.csv', encoding='utf-8', index=False)
```
