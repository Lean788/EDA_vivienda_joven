```python
salarios = pd.read_excel('./data/dataset1_salarios.xlsx', index_col=None, header=None)
salarios.head(15)
```

```python
salarios.columns
```

```python

salarios.dropna(thresh=9,inplace=True)

salarios.head()
```

```python
salarios.reset_index(drop=True ,inplace=True)
```

```python
salarios.columns
```

```python
# salarios.drop(['index'], axis = 1, inplace = True)
salarios
```

```python
salarios.info()
```

```python
salarios.iloc[0][1:] = salarios.iloc[0][1:].astype(int)
```

```python
salarios.columns = salarios.iloc[0]
```

```python
salarios.head()
```

```python

salarios.rename(columns={'TIME':'Paises', 2013.0: int(2013), 
                                     2014.0: int(2014), 
                                     2015.0: int(2015), 
                                     2016.0: int(2016), 
                                     2017.0: int(2017), 
                                     2018.0: int(2018), 
                                     2019.0: int(2019), 
                                     2020.0: int(2020), 
                                     2021.0: int(2021)}, inplace=True)
```

```python
salarios.head()
```

```python
salarios.drop([0], inplace=True)
salarios
```

```python
salarios.to_csv('./data/clean/salarios.csv', encoding='utf-8', index=False)
```
