```python
import string

def quitar_puntuacion(cadena):
    
    out = ''.join([i for i in cadena if i not in string.punctuation])
    return out
```

```python
for cadena in poblacion[2022]:
    quitar_puntuacion(cadena)

poblacion[2022]
```
