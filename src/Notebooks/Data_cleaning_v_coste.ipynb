## Dataset de Sobrecoste de la vivienda 
### Tasa de sobrecarga del costo de la vivienda por estado de tenencia: encuesta EU-SILC.
### Unidades: porcentaje

Fuente: Eurostat. Ref: ILC_LVHO07C

```python
# importaciones básicas
import numpy as np
import pandas as pd

# pd.set_option('max_rows', None)
pd.set_option('max_columns', None)
# pd.set_option('display.width', None)
# pd.set_option('display.max_colwidth', None)
```

```python
sobrecoste_v = pd.read_csv('../../src/data/dataset9_sobrecoste_v.csv', sep=';')
sobrecoste_v
```

```python
sobrecoste_v.info()
```

```python
sobrecoste_v.columns
```

```python
sobrecoste_v.index = sobrecoste_v['TIME']
```

```python
sobrecoste_v.drop(columns='TIME', inplace=True)
```

```python
sobrecoste_v_T = sobrecoste_v.T
```

```python
sobrecoste_v_T
```

```python
sobrecoste_v_T['France'].astype('float', errors='ignore')
```

```python
for column in sobrecoste_v_T:
    sobrecoste_v_T[column] = sobrecoste_v_T[column].astype('float',errors= 'ignore')
```

```python
sobrecoste_v_T.info()
```

```python
sobrecoste_v_T['France']
```

```python
sobrecoste_v_T['France'] = sobrecoste_v_T['France'].map({'16.2':float(16.2),
                                                        '16.2':float(16.2),
                                                        '15.8':float(15.8),
                                                        '14.9':float(14.9),
                                                        '16.5':float(16.5),
                                                        '14.7':float(14.7),
                                                        '14.9':float(14.9),
                                                        '17.7':float(17.7),
                                                        '18.2':float(18.2)})
```

```python
sobrecoste_v_T
```

```python
values = {'2017':14.7, '2018':14.9,'2021':'-'}
```

```python
sobrecoste_v_T['France'] = sobrecoste_v_T['France'].fillna(value=values)
```

```python
sobrecoste_v_T.tail(3)
```

```python
sobrecoste_v_T.to_csv('../data/clean/sobrecoste_v.csv', encoding='utf-8', index=False)
```
