```python
indicadores_transposed.rename(columns={'    Población residente (Personas)':'poblacion',
                                        '    Proporción de población de 15-64 años (Porcentaje)':'Prop_pob_15_64',
                                        '    Número total de hogares (Número)':'num_hogares',
                                        '    Proporción de viviendas vacías (Porcentaje)':'v_vacias',
                                        '    Renta neta media anual de los hogares (Euros)':'rnma_hogares',
                                        '    Renta neta media anual por habitante (Euros)':'rnma_hab'}, inplace=True)
```

```python
indicadores_transposed['poblacion'] = indicadores_transposed['poblacion'].astype('int32')
indicadores_transposed['num_hogares'] = indicadores_transposed['num_hogares'].astype('int32')
```

```python
sns.set(style="darkgrid")    
fig, ax = plt.subplots(figsize=(8, 5))    
g=sns.histplot(ax=ax,
             y=indicadores_transposed['num_hogares'], 
             x=indicadores_transposed.index,
             kde=True,
             bins=100, 
             data=indicadores_transposed);
ylabels = ['{:,.2f}'.format(y) + 'M' for y in g.get_yticks()/1000000]
g.set_yticklabels(ylabels);

# plt.xticks(rotation=-45, fontsize=10);
```
