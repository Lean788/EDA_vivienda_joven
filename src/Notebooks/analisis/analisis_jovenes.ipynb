### Definición de las variables:

- **Población**: Censo del Nº de personas residentes en España.
- **num_hogares**: Nº de viviendas principales en España.
- **precio_eur**: Precio del alquiler por metro cuadrado, en España.
- **num_hipotecas**: Nº de hipotecas firmadas, datos agregados anuales.
- **Total**: Porcentaje de número de hogares en regimen de alquiler a precio de mercado. 


### Conclusiones sobre este gráfico:

- La correlación se encuentre los valores: [0.16 - 0.91].
- La diagonal muestra una correlación de valor 1 (Correlación perfecta) porque es consigo misma.
- Encontramos **fuertes correlaciones** entre el `porcentaje de número de hogares en alquiler` y `el precio por metro cuadrado` del mismo. Aquí podemos ver que se refleja la realidad del mercado, ya que se puede verificar que a un cambio en el precio del alquiler, puede suponer una fuerte influencia sobre la elección a un hogar en regimen de alquiler.
- Otra **fuerte correlación** se puede apreciar entre el `porcentaje de número de hogares en alquiler` y el `número total de hipotecas firmadas`. Tiene sentido, ya que si aumentan o disminuye el número de hipotecas firmadas, hará que muchos cambien su actual régimen de tenencia en el hogar, pudiendo pasar de arrendatario a propietario.
- Otra **fuerte correlación** es entre el `total de la población en España` y el `precio del alquiler`. Siendo el alquiler el régimen más asequible, es de esperar que los jovenes inicien su etapa de independencia con estas opciones, antes de poder reunir las condiciones necesarias para poder adquirir una vivienda.
